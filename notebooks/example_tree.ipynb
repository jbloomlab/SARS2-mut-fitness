# Draw an example tree illustrating the process

```python
import io

import Bio.Phylo

import matplotlib
```

Define newick string for a tree:

```python
newick_str = "((((t1:1,t2:1)n1:1,(t3:1,t4:1)n2:1)n3:1),(t5:1,t6:1)n5:1)n4:1;"

tree = Bio.Phylo.read(io.StringIO(newick_str), "newick")

Bio.Phylo.draw(tree)
```
